```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def sort_data(file):
    '''
    Sort data in ascending order and save\n
    File: takes in the file path with extension. e.g: 'data/lewis.csv'\n 
    with different files feel free to adjust the name and dir of output file
    '''
    data = pd.read_csv(file)
    data['SN'] = data['Name'].str.extract(r'_(\d+)$').astype(int)
    data = data.sort_values(by=['SN']).drop(columns='SN')
    print('sorting successful, preview new data ...')
    print(data)
    print('saving sorted data ...')
    data.to_csv('../data/sorted_descriptors3.csv', index=False) 
    print('saved  data/ directory')
```

```python
sort_data('../data/PaDEL_descriptors3.csv')
```
