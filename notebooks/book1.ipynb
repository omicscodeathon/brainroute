```python
df_class = df_class.rename(columns={'compound_name':'name', 'SMILES':'smiles'})
df_class

df_molnet = df_molnet.rename(columns={'p_np': 'BBB'})
df_molnet
```

```python
import numpy as np
df_class['BBB'] = np.where(df_class['BBB+/BBB-'] == "BBB+", 1, 0)
```

```python
#cleaning b3bd data and combining dataframes 
#combining the classification and moleculenet  

#df_merged = pd.merge(df_class, df_molnet, on =['smiles', 'name', 'BBB'], how='outer')
#df_merged

names = pd.concat([df_class['name'], df_molnet['name']], ignore_index=True)

smiles = pd.concat([df_class['smiles'], df_molnet['smiles']], ignore_index=True)

BBB = pd.concat([df_class['BBB'], df_molnet['BBB']], ignore_index=True)
```

```python
df_combined = pd.DataFrame({
    'name': names,
    'smiles': smiles,
    'BBB': BBB
})
df_combined
```

```python
from rdkit import Chem

def canonicalize(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None
        # remove explicit hydrogens if you like:
        # mol = Chem.RemoveHs(mol)
        can = Chem.MolToSmiles(mol, isomericSmiles=True)
        return can
    except Exception:
        return None

df_combined['smiles_can'] = df_combined['smiles'].map(canonicalize)
df3 = df_combined.dropna(subset=['smiles_can'])
df3
```

```python
counts = df3.groupby('name')['smiles_can'].nunique() #number of smiles per name

# find names with conflicting labels (e.g., multiple logBB different)
print(counts)
```
