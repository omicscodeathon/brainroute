Dataset - Chembl, target search for - Serine/threonine-protein kinase mTOR

```python
datapath = "../data/mtor_inhibitors_approved26.tsv"

df = pd.read_csv(datapath, sep='\t')

df
```

## Retrieving SMILES

```python
df_required = df[['Parent Molecule ChEMBL ID', 'Smiles']].dropna()
df_required
```

## Loading models

```python
import joblib
from tensorflow.keras import models 

def KNN_model():
    model = joblib.load('../output/models/KNN_model.pkl')
    return model

def SVM_model():
    model = joblib.load('../output/models/SVM_model.pkl')
    return model

def RF_model():
    model = joblib.load('../output/models/RF_model.pkl')
    return model

def LR_model():
    model = joblib.load('../output/models/LR_model.pkl')
    return model

def XGB_model():
    model = joblib.load('../output/models/XGB_model.pkl')
    return model

def MLP_model():
    model = models.load_model('../output\models\MLP_model.h5')
    return model

def Scaler():
    scaler = joblib.load('../output/models/scaler.pkl')
    return scaler
```

## Running and comparing models

```python
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors, Crippen, Lipinski

def get_desc(smiles):
    mol = Chem.MolFromSmiles(smiles)
    descr = pd.DataFrame([Descriptors.CalcMolDescriptors(mol)])
    return descr
```

```python
test = df_required.iloc[0]
print(test['Smiles'])
print(get_desc(test['Smiles']))
```

```python
scaler = Scaler()
KNN_results = {}
knn = KNN_model()
for i in range(len(df_required)):
    n = df_required.iloc[i]
    smiles = n['Smiles']
    desc = get_desc(smiles)
    pred = knn.predict(scaler.transform(desc))
    confidence = knn.predict_proba(scaler.transform(desc))
    bbb_prediction = "BBB+" if pred[0] == 1 else "BBB-"
    conf_score = confidence[0][pred[0]] * 100
    result = [bbb_prediction, conf_score]
    KNN_results[n['Parent Molecule ChEMBL ID']] = result

    
KNN_results
```

```python
SVM_results = {}
svm = SVM_model()
for i in range(len(df_required)):
    n = df_required.iloc[i]
    smiles = n['Smiles']
    desc = get_desc(smiles)
    pred = svm.predict(scaler.transform(desc))
    confidence = svm.predict_proba(scaler.transform(desc))
    bbb_prediction = "BBB+" if pred[0] == 1 else "BBB-"
    conf_score = confidence[0][pred[0]] * 100
    result = [bbb_prediction, conf_score]
    SVM_results[n['Parent Molecule ChEMBL ID']] = result
    
SVM_results
```

```python
RF_results = {}
rf = RF_model()
for i in range(len(df_required)):
    n = df_required.iloc[i]
    smiles = n['Smiles']
    desc = get_desc(smiles)
    pred = rf.predict(scaler.transform(desc))
    confidence = rf.predict_proba(scaler.transform(desc))
    bbb_prediction = "BBB+" if pred[0] == 1 else "BBB-"
    conf_score = confidence[0][pred[0]] * 100
    result = [bbb_prediction, conf_score]
    RF_results[n['Parent Molecule ChEMBL ID']] = result

RF_results
```

```python
LR_results = {}
lr = LR_model()
for i in range(len(df_required)):
    n = df_required.iloc[i]
    smiles = n['Smiles']
    desc = get_desc(smiles)
    pred = lr.predict(scaler.transform(desc))
    confidence = lr.predict_proba(scaler.transform(desc))
    bbb_prediction = "BBB+" if pred[0] == 1 else "BBB-"
    conf_score = confidence[0][pred[0]] * 100
    result = [bbb_prediction, conf_score]
    LR_results[n['Parent Molecule ChEMBL ID']] = result
    
LR_results
```

```python
XGB_results = {}
xgb = XGB_model()
for i in range(len(df_required)):
    n = df_required.iloc[i]
    smiles = n['Smiles']
    desc = get_desc(smiles)
    pred = xgb.predict(scaler.transform(desc))
    confidence = xgb.predict_proba(scaler.transform(desc))
    bbb_prediction = "BBB+" if pred[0] == 1 else "BBB-"
    conf_score = confidence[0][pred[0]] * 100
    result = [bbb_prediction, conf_score]
    XGB_results[n['Parent Molecule ChEMBL ID']] = result
    
XGB_results
```

We can compare several models to validate our results. Notably, none of the currently approved clinical drug candidates targeting serine/threonine-protein kinases are capable of crossing the bloodbrain barrier. This highlights the significant challenge the barrier poses in the development of effective therapeutics.

# Joint model Comparison of results

```python
smiles = df_required['Smiles']
KNN = []
SVM = []
RF = []
LR = []
XGB = []

for smile in smiles:
    desc = get_desc(smile)
    KNN.append(knn.predict(scaler.transform(desc))[0])
    SVM.append(svm.predict(scaler.transform(desc))[0])
    RF.append(rf.predict(scaler.transform(desc))[0])
    LR.append(lr.predict(scaler.transform(desc))[0])
    XGB.append(xgb.predict(scaler.transform(desc))[0])

results = pd.DataFrame()
results['ChemBL_ID'] = df_required['Parent Molecule ChEMBL ID']
results['Smiles'] = smiles
results['XGB'] = XGB
results['KNN'] = KNN
results['SVM'] = SVM
results['RF'] = RF
results['LR'] = LR

```

```python
results
```

# Combined prediction from models

### Taking the most frequent prediction for molecules from all models

```python
results['Final_Prediction'] = 0
for i in range(len(results['Smiles'])):
    if (results.XGB.iloc[i] + results.KNN.iloc[i] + results.SVM.iloc[i] + results.RF.iloc[i] + results.LR.iloc[i]) >= 3:
        results['Final_Prediction'].iloc[i] = 1
        
results
```

```python
results.replace(1, 'BBB+', inplace=True)
results.replace(0, 'BBB-', inplace=True)
results
```
