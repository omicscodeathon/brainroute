```python
import pandas as pd 
```

# Importing Datasets

```python
combined_bbb = pd.read_csv('../data/combined_bbb_classification_no_descr.csv')
combined_bbb.head()
```

```python
combined_bbb.shape
```

```python
descr = pd.read_csv('../data/rdkit_descriptors.csv')
descr.head()
```

```python
descr.shape
```

```python
# check for duplicates
print(f'combined_bbb smile duplicates: {combined_bbb['smiles'].duplicated().sum()}')
print(f'descr smile duplicates: {descr['smiles'].duplicated().sum()}')
```

```python
# remove duplicates if any
combined_bbb = combined_bbb.drop_duplicates(subset='smiles', keep=False)
descr = descr.drop_duplicates(subset='smiles', keep=False)
```

```python
combined_bbb.shape, descr.shape
```

# merge descriptors with combined data

```python
combined_bbb_descr = pd.merge(combined_bbb, descr, on='smiles', how='left')
combined_bbb_descr.head()
```

```python
combined_bbb_descr.shape
```

```python
# save data to csv file
combined_bbb_descr.to_csv('../data/combined_bbb_classification.csv', index=False)
```
